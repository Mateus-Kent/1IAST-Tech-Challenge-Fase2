# 📊 EDA — Exploração dos Dados Brutos

**Tech Challenge – Fase 2 | PosTech FIAP**

Este notebook explora os dados brutos baixados do [Base dos Dados](https://basedosdados.org/) antes de qualquer transformação.

O objetivo é entender:
- O que cada tabela contém
- Volume de dados
- Qualidade: nulos, tipos, duplicatas
- Chaves de relacionamento entre as tabelas

```python
import pandas as pd
import os
from pathlib import Path

RAW_DIR = Path("..") / "data" / "raw"

print("Arquivos disponíveis em data/raw/:")
for f in sorted(RAW_DIR.glob("*.csv")):
    size_kb = f.stat().st_size / 1024
    print(f"  {f.name:<45} {size_kb:>8.1f} KB")
```

## 1. Carregando as tabelas

```python
# Carrega todas as tabelas
meta_brasil    = pd.read_csv(RAW_DIR / "meta_alfabetizacao_brasil.csv")
meta_uf        = pd.read_csv(RAW_DIR / "meta_alfabetizacao_uf.csv")
meta_municipio = pd.read_csv(RAW_DIR / "meta_alfabetizacao_municipio.csv")
indicador_uf   = pd.read_csv(RAW_DIR / "indicador_uf.csv")
indicador_mun  = pd.read_csv(RAW_DIR / "indicador_municipio.csv")

tabelas = {
    "meta_brasil":    meta_brasil,
    "meta_uf":        meta_uf,
    "meta_municipio": meta_municipio,
    "indicador_uf":   indicador_uf,
    "indicador_mun":  indicador_mun,
}

print(f"{'Tabela':<20} {'Linhas':>8} {'Colunas':>8}")
print("-" * 40)
for nome, df in tabelas.items():
    print(f"{nome:<20} {df.shape[0]:>8,} {df.shape[1]:>8}")
```

## 3. Meta Alfabetização por UF

## 7. Relacionamento entre as tabelas

Verificamos se os municípios presentes em `meta_municipio` existem também em `indicador_municipio`.

```python
ids_meta = set(meta_municipio["id_municipio"].unique())
ids_ind  = set(indicador_mun["id_municipio"].unique())

so_em_meta = ids_meta - ids_ind
so_em_ind  = ids_ind  - ids_meta
em_ambos   = ids_meta & ids_ind

print(f"Municípios em meta_municipio:      {len(ids_meta):,}")
print(f"Municípios em indicador_municipio: {len(ids_ind):,}")
print(f"Em ambas as tabelas:               {len(em_ambos):,}")
print(f"Só em meta (sem indicador):        {len(so_em_meta):,}")
print(f"Só em indicador (sem meta):        {len(so_em_ind):,}")
```

## 8. Duplicatas

```python
for nome, df in tabelas.items():
    dupes = df.duplicated().sum()
    print(f"{nome:<20}: {dupes} duplicatas")
```

## 9. Resumo — Pontos de atenção para a Silver Layer

Com base na EDA, os principais pontos a tratar na Silver são:

| Problema | Tabela | Ação na Silver |
|---|---|---|
| Colunas com muitos nulos | `meta_municipio`, `indicador_municipio` | Manter flag `dados_ausentes` |
| Tipos como `object` devem ser numéricos | Todas | Converter com `pd.to_numeric` |
| Municípios sem correspondência | `meta` vs `indicador` | Investigar e documentar |
| Coluna `rede` com valores diferentes | Algumas tabelas | Padronizar: `Pública`, `Privada`, `Municipal` |
